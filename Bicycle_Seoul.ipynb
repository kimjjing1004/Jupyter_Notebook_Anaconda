```python
import pandas as pd
import glob
```

```python
path = './data/Bicycle/20'
all_files = glob.glob(path + "/*.csv")
ls = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    ls.append(df)
total_menu = pd.concat(ls, axis=0, ignore_index=True)
```

```python
total_menu
```

```python
data = total_menu[total_menu['대여일자'] < '2020-01-01']
data
```

```python
# csv 저장(to_csv)
data.to_csv('data/Bicycle/20/Bicycle_19_12.csv', sep=',')
```
