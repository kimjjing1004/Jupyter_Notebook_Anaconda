```python
import pandas as pd
```

```python
Bicycle_Seoul = pd.read_csv('data/Bicycle/sample.csv', encoding='utf-8')

Bicycle_Seoul
```

```python
grouped_Bicycle_Seoul1 = Bicycle_Seoul.groupby(['연월', '연령대코드'])['사용시간(분)']

grouped_Bicycle_Seoul1
```

```python
grouped_Bicycle_Seoul1.agg(['mean','sum'])
```
